```python
import numpy as np

arr = np.array([1,2,3,4,])
arr #array([1, 2, 3, 4, 5])

type(arr) #numpy.ndarray
arr.dtype #dtype('int64')

arr[0]= 11
arr #array([11,  2,  3,  4,  5])

arr.ndim #1

arr.shape #(5,)

arr.size #5
```

```python
frr = np.array([1.2,2.3,3.4,4.5])

arr + frr #array([12.2,  4.3,  6.4,  8.5])

arr / frr #array([9.16666667, 0.86956522, 0.88235294, 0.88888889])

arr ** frr #array([ 17.76933693,   4.92457765,  41.8998305 , 512.        ])
```

Slicing works same as in python but has more functionality due to multi-dimensional.


```python
a = np.array([
    [0, 1, 2, 3, 4, 5],
    [10, 11, 12, 13, 14, 15],
    [20, 21, 22, 23, 24, 25],
    [30, 31, 32, 33, 34, 35],
    [40, 41, 42, 43, 44, 45],
    [50, 51, 52, 53, 54, 55]
])

# 1. Slicing a specific row and column range
print(a[0, 3:5])  # Output: array([3, 4])

# 2. Slicing from a specific row and column onwards
print(a[4:, 4:])  # Output: array([[44, 45], [54, 55]])

# 3. Extracting a specific column
print(a[:, 2])  # Output: array([ 2, 12, 22, 32, 42, 52])

# 4. Strided slicing
print(a[2::2, ::2])  # Output: array([[20, 22, 24], [40, 42, 44]])
```

```python
"""
Filter Image
------------

Read in the "dc_metro" image and use an averaging filter
to "smooth" the image. Use a "5 point stencil" where
you average the current pixel with its neighboring pixels::

     0  0  0  0  0
     0  x  0  0  0
     0  x  x  x  0
     0  0  x  0  0
     0  0  0  0  0

Plot the image, the smoothed image, and the difference between
the two.

Bonus
-----

Re-filter the image by passing the result image through the
filter 50 times and plot the resulting image.

See :ref:`filter-image-solution`.
"""

import matplotlib.pyplot as plt

img = plt.imread('metro.png')

top = img[0:-2, 1:-1]
left = img[1:-1, 0:-2]
right = img[1:-1, 2:]
bottom = img[2:, 1:-1]
center = img[1:-1, 1:-1]

blurred_img = (top + left + right + bottom + center) / 5

top = blurred_img[0:-2, 1:-1]
left = blurred_img[1:-1, 0:-2]
right = blurred_img[1:-1, 2:]
bottom = blurred_img[2:, 1:-1]
center = blurred_img[1:-1, 1:-1]

blurred_twice_img = (top + left + right + bottom + center) / 5


plt.imshow(img, cmap=plt.cm.hot)
plt.figure()
plt.imshow(blurred_img)
plt.figure()
plt.imshow(blurred_twice_img)
plt.show()
```
