```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
import matplotlib.pyplot as plt
```

```python
tips.tip.plot(kind='hist')
plt.show()
```

```python
cts = tips['smoker'].value_counts()
cts
```

```python
cts.plot(kind='bar')
```

```python
sns.countplot(x='smoker', data= tips, hue='smoker')
#better plotting function and UI as well for visualization
```

```python
sns.displot(tips.total_bill, kde = True)
#discrete plot
```

```python
# sns.lmplot(x='total_bill', y='tip', data= tips, hue ='sex', fit_reg= False)
sns.lmplot(x='total_bill', y='tip', data= tips, hue ='sex', fit_reg= False, col='smoker')
#line regression plot
```

```python
sns.lmplot(x='total_bill', y='tip', data= tips, hue ='sex', fit_reg= False, col='smoker', row='day')
```

```python
facet = sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
facet.map(plt.scatter, 'total_bill', 'tip')
```

```python
fig, (ax) =plt.subplots(1,1)
ax.scatter(tips.tip, tips.total_bill)
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.scatter(tips.tip, tips.total_bill)
ax2.hist(tips.total_bill)
```
